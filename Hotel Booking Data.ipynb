```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("hotel_booking_data.csv")
```

```python
df
```

```python
# 1.What are the top 5 most common country codes in the 
```

```python
df["country"].value_counts()[:5]
```

```python
# 2.What is the name of the person who paid the highest ADR (average daily rate)? How much was their ADR?
```

```python
df.sort_values('adr', ascending=False)[['adr', 'name']].iloc[0]
```

```python
# 3.What is the average (mean) number of nights for a stay across the entire data set?
```

```python
df['total_stay_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
round(df['total_stay_nights'].mean(),2)
```

```python
# 4.What is the average total cost for a stay in the dataset?
```

```python
round(df['adr'].mean(),2)
```

```python
# 5.What are the names and emails of people who made 5 "Special Requests"?
```

```python
df[df['total_of_special_requests'] == 5][['name', 'email']]
```

```python
# 6.What percentage of hotel stays were classified as "repeat guests"?
```

```python
round(100*sum(df['is_repeated_guest']==1) / len(df),2)
```

```python
# 7.What are the top 5 most common last name in the dataset?
```

```python
df['name'].apply(lambda name : name.split()[-1]).value_counts()[:5]
```

```python
# 8.What are the names of the people who had booked the most number children and babies for their stay?
```

```python
df['total_kids'] = df['babies'] + df['children']
df.sort_values('total_kids', ascending=False)[['name', 'total_kids']]
```

```python
# 9.What are the top 3 most common area code in the phone numbers? 
```

```python
df['phone-number'].apply(lambda num : num[:3]).value_counts()
```

```python
# 10.How many arrivals took place between the 1st and the 15th of the month (inclusive of 1 and 15) ?
```

```python
df['arrival_date_day_of_month'].apply(lambda day: day in range(1,16)).sum()
```
